```python
results_10 = results[1:][[ 
       'Votes_10', 'Voters_10', 'Votes_10_dem_prim', 'Votes_10_rep_prim']]
results_14 = results[1:][[ 
       'Votes_14', 'Voters_14', 'Votes_14_dem_prim', 'Votes_14_rep_prim']]
results_18 = results[1:][[ 
       'Votes_14', 'Voters_14', 'Votes_14_dem_prim', 'Votes_14_rep_prim']]
for result in [results_10, results_14, results_18]:
    result.columns = ["Votes", "Voters", "Votes_Dem_Prim", "Votes_Rep_Prim"]
results_22 = results[1:][[ 
       'Voters_22', 'Votes_22_dem_prim', 'Votes_22_rep_prim']]
results_22.columns = ["Voters", "Votes_Dem_Prim", "Votes_Rep_Prim"]
results_10.head()
```

<p> Now that all my dataframes have the same row indexes, I can merge them by this index! I need to merge all my datasets horizontally & vertically (I will distinguish between years vertically) <p>

```python
#do work here
data = []
res = [results_10, results_14, results_18, results_22]
age = [age_10, age_14, age_18, age_20]
income = [income_10, income_14, income_18, income_20]
race = [race_10, race_14, race_18, race_20]
for i in range(0,4):
    a = res[i].join(race[i].join(income[i].join(age[i])))
    a.drop('Total:', axis=1, inplace=True)
    a.drop('Total population', axis=1, inplace=True)
    data.append(a)
data10 = data[0]
data14 = data[1]
data18 = data[2]
data22 = data[3]
data22.head()
```

```python
def fix_cols(data):
    #ensuring all the datasets have the same columns when they merge
    data.columns = ['Votes', 'Voters', 'Votes_Dem_Prim', 'Votes_Rep_Prim',
       'White', 'Black or African American',
       'American Indian and Alaska Native', 'Asian',
       'Native Hawaiian Pacific Islander',
       'Other', 'Two Races',
       'Two Races Including Other',
       'Three or More Races',
       'Median Income', 'Median Age',
       'Sex Ratio']
    #also making all values within each column numeric for data analysis
    data[data.columns] = data[data.columns].apply(pd.to_numeric, errors='coerce')
    #combining race data for all "other" categories for better interpretation (also get rid of duplicate categories)
    data["Other Race"] = data['Other'] + data['Two Races']
    return data[['Votes', 'Voters', 'Votes_Dem_Prim', 'Votes_Rep_Prim',
       'White', 'Black or African American',
       'American Indian and Alaska Native', 'Asian',
       'Native Hawaiian Pacific Islander',
       'Other Race',
       'Median Income', 'Median Age',
       'Sex Ratio']]
data10 = fix_cols(data10)
data14 = fix_cols(data14)
data18 = fix_cols(data18)

#2022 data does not have a votes column so function does not apply
data22[data22.columns] = data22[data22.columns].apply(pd.to_numeric, errors='coerce') 
data22.columns = ['Voters', 'Votes_Dem_Prim', 'Votes_Rep_Prim',
       'White', 'Black or African American',
       'American Indian and Alaska Native', 'Asian',
       'Native Hawaiian Pacific Islander',
       'Other', 'Two Races',
       'Two Races Including Other',
       'Three or More Races',
       'Median Income', 'Median Age',
       'Sex Ratio']
data22["Other Race"] = data22['Other'] + data22['Two Races']
data22 = data22[['Voters', 'Votes_Dem_Prim', 'Votes_Rep_Prim',
       'White', 'Black or African American',
       'American Indian and Alaska Native', 'Asian',
       'Native Hawaiian Pacific Islander',
       'Other Race',
       'Median Income', 'Median Age',
       'Sex Ratio']]
data22.head()
```
